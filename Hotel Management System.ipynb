## Class Room

```python
class Room:
    # Constructor
    def __init__(self , room_id , room_type , price_per_night):
        self.__room_id = room_id
        self.__room_type = None
        self.__price_per_night = None
        self.__is_available = True
        # sets values
        self.set_room_type(room_type)
        self.set_price_per_night(price_per_night)
        
    # Setters
    def set_room_id(self , room_id):
        self.__room_id = room_id
    
    def set_room_type(self , room_type):
        if room_type.lower() in ["single" , "double" , "suite"]:
            self.__room_type = room_type
        else:
            raise ValueError("Invalid Room type")
        
    def set_price_per_night(self , price_per_night):
        if price_per_night > 0:
            self.__price_per_night = price_per_night
        else:
            raise ValueError("Invalid Price per night")
        
    def set_room_status(self , is_available):
        if isinstance(is_available , bool):
            self.__is_available = is_available
        else:
            raise ValueError("Room status must be True or False")
        
    # Getters
    def get_room_id(self):
        return self.__room_id
    
    def get_room_type(self):
        return self.__room_type
    
    def get_price_per_night(self):
        return self.__price_per_night
    
    def get_room_status(self):
        return self.__is_available
    
    # Methods
    def get_details(self):
        return f"Room id = {self.get_room_id()} , it's type: {self.get_room_type()} , price per night = {self.get_price_per_night()} and it's availability status is : {self.get_room_status()}"
    
    def mark_as_booked(self):
        self.set_room_status(False)
        
    def mark_as_available(self):
        self.set_room_status(True)
```

## Class Guest

```python
# Parent Class
class Person:
    def __init__(self , id , name):
        # protected attributes 
        self._id = id
        self._name = name
        
    def get_details(self):
        return f"id = {self._id} and his name = {self._name}"
```

```python
# Child Class
class Guest(Person):
    # Constructor
    def __init__(self , guest_id , guest_name ,guest_phone , guest_email):
        super().__init__(guest_id , guest_name)
        self.__guest_phone = None
        self.__guest_email = None
        # set values
        self.set_guest_phone(guest_phone)
        self.set_guest_email(guest_email)
        
    # Setters
    def set_guest_id(self , guest_id):
        self._id = guest_id
        
    def set_guest_name(self , guest_name):
        self._name = guest_name
        
    def set_guest_phone(self , guest_phone):
        if guest_phone.isdigit() and len(guest_phone) == 11:
            self.__guest_phone = guest_phone
        else:
            raise ValueError("Invalid Phone number")
        
    def set_guest_email(self , guest_email):
        if "@" in guest_email:
            self.__guest_email = guest_email
        else:
            raise ValueError("Invalid Email Format")
        
    # Getters
    def get_guest_id(self):
        return self._id
    
    def get_guest_name(self):
        return self._name
    
    def get_guest_phone(self):
        return self.__guest_phone
    
    def get_guest_email(self):
        return self.__guest_email
    
    # Methods
    def get_details(self):
        return f"Guest with id {self.get_guest_id()} his name {self.get_guest_name()} his phone number: {self.get_guest_phone()} , his email: {self.get_guest_email()}"
```

## Class Reservation

```python
class Reservation:
    # Constructor
    def __init__(self , reservation_id , guest , room , check_in_date , check_out_date):
        self.__reservation_id = reservation_id
        self.__guest = None
        self.__room = None
        self.__check_in_date = None
        self.__check_out_date = None
        self.__total_price = 0
        self.__status = "Pending"
        
        # set values 
        self.set_guest(guest)
        self.set_room(room)
        self.set_check_in_date(check_in_date)
        self.set_check_out_date(check_out_date)
        
    # Setters
    def set_reservation_id(self , reservation_id):
        self.__reservation_id = reservation_id
        
    def set_check_in_date(self , check_in_date):
        if check_in_date >= date.today():
            self.__check_in_date = check_in_date
        else:
            raise Exception("Invalid Check in date")
        
    def set_check_out_date(self , check_out_date):
        if self.__check_in_date is None:
            raise Exception("Set check-in date first")
        if check_out_date > self.__check_in_date:
            self.__check_out_date = check_out_date
        else:
            raise Exception("Invalid Check out date")
        
    def set_status(self , status):
        if status.lower().strip() in ["cancelled" , "confirmed"]:
            self.__status = status
        else:
            raise ValueError("Invalid Status")
        
    def set_room(self , room):
        if isinstance(room , Room):
            self.__room = room
        else:
            raise TypeError("room must be a Room object")
        
    def set_guest(self , guest):
        if isinstance(guest , Guest):
            self.__guest = guest
        else:
            raise TypeError("guest must be a Guest object")
        
    def set_total_price(self , price):
        if price > 0:
            self.__total_price = price
        else:
            raise ValueError("Invalid Price")
        
    # Getters
    def get_reservation_id(self):
        return self.__reservation_id

    def get_guest(self):
        return self.__guest

    def get_room(self):
        return self.__room

    def get_check_in_date(self):
        return self.__check_in_date

    def get_check_out_date(self):
        return self.__check_out_date

    def get_total_price(self):
        return self.__total_price

    def get_status(self):
        return self.__status

    # Methods
    def calculate_price(self):
        if self.get_check_out_date() is None or self.get_check_in_date() is None:
            return 0
        
        number_of_days = (self.get_check_out_date() - self.get_check_in_date()).days
        return number_of_days * self.get_room().get_price_per_night()
    
    def confirm_booking(self):
        if self.get_room().get_room_status() == True:
            self.set_status("confirmed")
            self.get_room().mark_as_booked()
            self.set_total_price(self.calculate_price())
            return "Confirm booking successfully"
        
        raise Exception("Room is not available")
    
    def cancel_booking(self):
        if self.get_status().lower() in ["pending" , "confirmed"]:
            self.set_status("cancelled")
            self.get_room().mark_as_available()
            return "Reservation cancelled successfully."
        
        raise Exception("Cannot cancel this reservation.")
```

## Hotel abstract class

```python
class HotelAbstraction(ABC):
    # Room methods
    @abstractmethod
    def add_room(self , room):
        pass
    
    @abstractmethod
    def delete_room(self , room_id):
        pass
    
    @abstractmethod
    def search_room(self ,room_id):
        pass

    @abstractmethod
    def update_room(self,room_id , new_room_type , new_price_per_night , new_status):
        pass
    
    # Guest methods
    @abstractmethod
    def add_guest(self , guest):
        pass
    
    @abstractmethod
    def delete_guest(self ,guest_id):
        pass
    
    @abstractmethod
    def search_guest(self ,guest_id):
        pass

    @abstractmethod
    def  update_guest(self,guest_id ,new_name , new_phone , new_email):
        pass
    
    # Reservation methods
    @abstractmethod
    def make_reservation(self ,reservation):
        pass
    
    @abstractmethod
    def delete_reservation(self ,reservation_id):
        pass

    @abstractmethod
    def update_reservation(self, reservation_id, new_status=None, new_check_out_date=None):
        pass
    
    @abstractmethod
    def search_reservation(self , reservation_id):
        pass
```

```python
class Hotel(HotelAbstraction):
    # Class attribute
    hotel_name = "Bassam Hotel"
    
    # Room methods
    def add_room(self,room):
        if os.path.exists(room_file_path):
            with open(room_file_path , "r") as file:
                room_ids = []
                for line in file.readlines():
                    data = line.strip().split(" , ")
                    room_ids.append(int(data[0].strip()))
            
            with open(room_file_path , "a") as f:
                if room.get_room_id() not in room_ids:
                    f.write(f"{room.get_room_id()} , {room.get_room_type()} , {room.get_price_per_night()} , {room.get_room_status()}\n")
                    return "Room added successfully"
                else:
                    return "Room with this id is found please try again"
        else:
            return "Rooms file not found"      
        
        
    def delete_room(self,room_id):
        if os.path.exists(room_file_path):
            with open(room_file_path , "r") as f:
                lines = f.readlines()
                
            room_found = False
                
            with open(room_file_path , "w") as file:
                for line in lines:
                    data = line.strip().split(" , ")
                    id = data[0]
                    room_type = data[1]
                    price_per_night = data[2]
                    room_status = data[3]
                    
                    if int(id)== room_id:
                        room_found = True
                    else:    
                        file.write(f"{id} , {room_type} , {price_per_night} , {room_status}\n")
                
                if room_found:
                    return "Room deleted successfully"
                return "Room not found"
        else:
            return "Rooms file not found" 
        
        
    def search_room(self ,room_id):
        if os.path.exists(room_file_path):
            with open(room_file_path , "r") as f:
                lines = f.readlines()
                room_found = False
                
                for line in lines:
                    data = line.strip().split(" , ")
                    id = data[0]
        
                    if int(id)== room_id:
                        room_found = True
                        room_type = data[1]
                        price_per_night = data[2]
                        room_status = data[3]
                        break
                
                if room_found:
                    if room_status.strip().lower() == "true":
                        status = "Available"
                    else:
                        status = "Not Available"
                    return f"Room with id {id} it's type : {room_type} ,status {status} and price per night = {price_per_night}"
                
                return "Room not found"
        else:
            return "Rooms file not found"  
        
        
    def update_room(self,room_id , new_room_type , new_price_per_night , new_status):
        if os.path.exists(room_file_path):
            with open(room_file_path , "r") as f:
                lines = f.readlines()
                room_found = False
            
            if (new_price_per_night > 0) and (new_room_type in ["Single" , "Double" , "Suite"]) and (isinstance(new_status,bool)):
                with open(room_file_path , "w") as file:
                    for line in lines:
                        data = line.strip().split(" , ")
                        id = data[0]
                        
                        if int(id)== room_id:
                            room_found = True
                            file.write(f"{id} , {new_room_type} , {new_price_per_night} , {new_status}\n")
                            
                        else:    
                            file.write(line)
                    
                    if room_found:
                        return "Room updated successfully"
                    return "Room not found"
            else:
                return "Error in new inputs"
        else:
            return "Rooms file not found" 
            
        
    # Guest Methods
    def add_guest(self , guest):
        if os.path.exists(guest_file_path):
            with open(guest_file_path , "r") as file:
                guest_ids = []
                for line in file.readlines():
                    data = line.strip().split(" , ")
                    guest_ids.append(int(data[0].strip()))
            
            with open(guest_file_path , "a") as f:
                if guest.get_guest_id() not in guest_ids:
                    f.write(f"{guest.get_guest_id()} , {guest.get_guest_name()} , {guest.get_guest_phone()} , {guest.get_guest_email()}\n")
                    return "Guest added successfully"
                else:
                    return "Guest with this id is found please try again"
        else:
            return "Guests file not found"   
    
    
    def delete_guest(self ,guest_id):
        if os.path.exists(guest_file_path):
            with open(guest_file_path , "r") as f:
                lines = f.readlines()
                
            guest_found = False
                
            with open(guest_file_path , "w") as file:
                for line in lines:
                    data = line.strip().split(" , ")
                    id = data[0]
                    guest_name = data[1]
                    guest_phone = data[2]
                    guest_email = data[3]
                    
                    if int(id)== guest_id:
                        guest_found = True
                    else:    
                        file.write(f"{id} , {guest_name} , {guest_phone} , {guest_email}\n")
                
                if guest_found:
                    return "Guest deleted successfully"
                return "Guest not found"
        else:
            return "Guest file not found" 


    def search_guest(self ,guest_id):
        if os.path.exists(guest_file_path):
            with open(guest_file_path , "r") as f:
                lines = f.readlines()
                guest_found = False
                
                for line in lines:
                    data = line.strip().split(" , ")
                    id = data[0]
        
                    if int(id)== guest_id:
                        guest_found = True
                        guest_name = data[1]
                        guest_phone = data[2]
                        guest_email = data[3]
                        break
                
                if guest_found:
                    return f"Guest with id {id} his name : {guest_name} , phone number {guest_phone} and email = {guest_email}"
                
                return "Guest not found"
        else:
            return "Guest file not found"  
    
    
    def update_guest(self,guest_id ,new_name , new_phone , new_email):
        if os.path.exists(guest_file_path):
            with open(guest_file_path , "r") as f:
                lines = f.readlines()
                
            guest_found = False
            
            if (new_phone.isdigit() and len(new_phone) == 11) and ("@" in new_email):
                with open(guest_file_path , "w") as file:
                    for line in lines:
                        data = line.strip().split(" , ")
                        id = data[0]
                        guest_name = data[1]
                        guest_phone = data[2]
                        guest_email = data[3]
                        
                        if int(id)== guest_id:
                            guest_found = True
                            file.write(f"{id} , {new_name} , {new_phone} , {new_email}\n")
                            
                        else:    
                            file.write(line)
                    
                    if guest_found:
                        return "Guest updated successfully"
                    return "Guest not found"
            else:
                return "Error in new inputs"
        else:
            return "Guest file not found" 

    # Get room price
    def get_room_price(self, room_id):
        if os.path.exists(room_file_path):
            with open(room_file_path, "r") as f:
                for line in f:
                    data = line.strip().split(" , ")
                    if int(data[0]) == room_id:
                        return float(data[2])
        return None
    
    
    # Reservation methods
    def make_reservation(self ,reservation):
        if os.path.exists(reservation_file_path):
            with open(reservation_file_path , "r") as file:
                reservation_ids = []
                for line in file.readlines():
                    data = line.strip().split(" , ")
                    reservation_ids.append(int(data[0].strip()))
                    
            guest_check = self.search_guest(reservation.get_guest().get_guest_id())
            room_check = self.search_room(reservation.get_room().get_room_id())
            
            with open(reservation_file_path , "a") as f:
                if reservation.get_reservation_id() not in reservation_ids :
                    if ("not found" not in guest_check.lower() ) and ("not found" not in room_check.lower()):
                        room_price = self.get_room_price(reservation.get_room().get_room_id())
                        days = (reservation.get_check_out_date() - reservation.get_check_in_date()).days
                        total_price = days * room_price

                        f.write(f"{reservation.get_reservation_id()} , {reservation.get_room().get_room_id()} , {reservation.get_guest().get_guest_id()} , {reservation.get_check_in_date()} , {reservation.get_check_out_date()} , {total_price} , {reservation.get_status()}\n")
                        return "Reservation added successfully"
                    else:
                        return "Guest or Room not found."
                else:
                    return "Reservation with this id is found please try again"
                
        else:
            return "Reservation file not found"
    
    
    def delete_reservation(self ,reservation_id):
        if os.path.exists(reservation_file_path):
            with open(reservation_file_path , "r") as f:
                lines = f.readlines()
                
            reservation_found = False
                
            with open(reservation_file_path , "w") as file:
                for line in lines:
                    data = line.strip().split(" , ")
                    id = data[0]
                    room_id = data[1]
                    guest_id = data[2]
                    check_in_date = data[3]
                    check_out_date = data[4]
                    total_price = data[5]
                    status = data[6]
                    
                    if int(id)== reservation_id:
                        reservation_found = True
                    else:    
                        file.write(f"{id} , {room_id} , {guest_id} , {check_in_date} , {check_out_date} , {total_price} , {status}\n")
                
                if reservation_found:
                    return "Reservation deleted successfully"
                return "Reservation not found"
        else:
            return "Reservation file not found"
        
        
    def search_reservation(self , reservation_id):
        if os.path.exists(reservation_file_path):
            with open(reservation_file_path , "r") as f:
                lines = f.readlines()
                reservation_found = False
                
                for line in lines:
                    data = line.strip().split(" , ")
                    id = data[0]
        
                    if int(id)== reservation_id:
                        reservation_found = True
                        room_id = data[1]
                        guest_id = data[2]
                        check_in_date = data[3]
                        check_out_date = data[4]
                        total_price = data[5]
                        status = data[6]
                        break
                
                if reservation_found:
                    return f"Reservation with id {id} room id : {room_id} , guest id: {guest_id} , check in date: {check_in_date} , check out date: {check_out_date} it's status = {status} and total price = {total_price}"
                
                return "Reservation not found"
        else:
            return "Reservations file not found" 
        

    def update_reservation(self, reservation_id, new_status=None, new_check_out_date=None):
        if os.path.exists(reservation_file_path):
            with open(reservation_file_path, "r") as f:
                lines = f.readlines()

            reservation_found = False

            with open(reservation_file_path, "w") as file:
                for line in lines:
                    data = line.strip().split(" , ")

                    id = int(data[0])
                    room_id = data[1]
                    guest_id = data[2]
                    check_in_date = data[3]
                    check_out_date = data[4]
                    total_price = data[5]
                    status = data[6]

                    if id == reservation_id:
                        reservation_found = True
                        
                        # update status
                        if new_status is not None:
                            if not (isinstance(new_status , str) and new_status.strip().lower() in ["confirmed", "cancelled"]):
                                return "Invalid status"
                            
                            status = new_status.strip().lower()
                            
                        check_in_date = datetime.strptime(check_in_date, "%Y-%m-%d").date()
                        check_out_date = datetime.strptime(check_out_date ,"%Y-%m-%d" ).date()
                        
                        # update check_out_date
                        if new_check_out_date is not None:
                            new_check_out_date = datetime.strptime(new_check_out_date , "%Y-%m-%d").date()
                            
                            if new_check_out_date > check_in_date:
                                check_out_date = new_check_out_date
                            else:
                                return "Invalid check-out date"
                            
                        file.write(f"{id} , {room_id} , {guest_id} , {str(check_in_date)} , {str(check_out_date)} , {total_price} , {status}\n")

                    else:
                        file.write(line)

            if reservation_found:
                return "Reservation updated successfully"
            else:
                return "Reservation not found"
        else:
            return "Reservation file not found"
    
```

```python
# ===================== PLACEHOLDER HELPER =====================
def add_placeholder(entry, text):
    entry.insert(0, text)
    entry.config(fg="gray")

    def on_focus_in(event):
        if entry.get() == text:
            entry.delete(0, tk.END)
            entry.config(fg="black")

    def on_focus_out(event):
        if entry.get() == "":
            entry.insert(0, text)
            entry.config(fg="gray")

    entry.bind("<FocusIn>", on_focus_in)
    entry.bind("<FocusOut>", on_focus_out)


# ===================== DASHBOARD =====================
class Dashboard:
    def __init__(self, root, hotel):
        self.root = root
        self.hotel = hotel

        self.root.title("Hotel Management System")
        self.root.geometry("950x550")
        self.root.configure(bg="#0f172a")

        # ================= SIDEBAR =================
        self.sidebar = tk.Frame(root, bg="#1e293b", width=220)
        self.sidebar.pack(side="left", fill="y")

        tk.Label(self.sidebar, text="🏨 HOTEL SYSTEM",
                bg="#1e293b", fg="white",
                font=("Arial", 14, "bold")).pack(pady=25)

        tk.Button(self.sidebar, text="🛏 Rooms",
                bg="#334155", fg="white",
                command=self.show_rooms, width=20).pack(pady=10)

        tk.Button(self.sidebar, text="👤 Guests",
                bg="#334155", fg="white",
                command=self.show_guests, width=20).pack(pady=10)

        tk.Button(self.sidebar, text="📅 Reservations",
                bg="#334155", fg="white",
                command=self.show_reservations, width=20).pack(pady=10)

        # ================= MAIN AREA =================
        self.main = tk.Frame(root, bg="#0f172a")
        self.main.pack(side="right", expand=True, fill="both")

        self.show_rooms()

    def clear(self):
        for w in self.main.winfo_children():
            w.destroy()

    # =========================================================
    #                         ROOMS
    # =========================================================
    def show_rooms(self):
        self.clear()

        tk.Label(self.main, text="🛏 Rooms Management",
                bg="#0f172a", fg="white",
                font=("Arial", 16, "bold")).pack(pady=15)

        tk.Label(self.main, text="Room ID", fg="white", bg="#0f172a").pack()
        id_entry = tk.Entry(self.main)
        id_entry.pack()
        add_placeholder(id_entry, "Enter Room ID")

        tk.Label(self.main, text="Room Type", fg="white", bg="#0f172a").pack()
        type_entry = ttk.Combobox(self.main,
                                values=["Single", "Double", "Suite"],
                                state="readonly")
        type_entry.set("Single")
        type_entry.pack()

        tk.Label(self.main, text="Price per Night", fg="white", bg="#0f172a").pack()
        price_entry = tk.Entry(self.main)
        price_entry.pack()
        add_placeholder(price_entry, "Enter Price")

        tk.Label(self.main, text="Availability", fg="white", bg="#0f172a").pack()
        status_entry = ttk.Combobox(self.main,
                                values=["True", "False"],
                                state="readonly")
        status_entry.set("True")
        status_entry.pack()

        # ===== Actions =====
        def add():
            try:
                room = Room(int(id_entry.get()),
                            type_entry.get(),
                            float(price_entry.get()))
                messagebox.showinfo("Result", self.hotel.add_room(room))
            except Exception as e:
                messagebox.showerror("Error", str(e))

        def delete():
            messagebox.showinfo("Result",
                                self.hotel.delete_room(int(id_entry.get())))

        def update():
            try:
                status = True if status_entry.get() == "True" else False
                result = self.hotel.update_room(
                    int(id_entry.get()),
                    type_entry.get(),
                    float(price_entry.get()),
                    status
                )
                messagebox.showinfo("Result", result)
            except Exception as e:
                messagebox.showerror("Error", str(e))

        def search():
            messagebox.showinfo("Result",
                                self.hotel.search_room(int(id_entry.get())))

        tk.Button(self.main, text="Add Room", bg="#22c55e", fg="white", command=add).pack(pady=4)
        tk.Button(self.main, text="Delete Room", bg="#ef4444", fg="white", command=delete).pack(pady=4)
        tk.Button(self.main, text="Update Room", bg="#3b82f6", fg="white", command=update).pack(pady=4)
        tk.Button(self.main, text="Search Room", bg="#eab308", fg="black", command=search).pack(pady=4)

    # =========================================================
    #                         GUESTS
    # =========================================================
    def show_guests(self):
        self.clear()

        tk.Label(self.main, text="👤 Guests Management",
                bg="#0f172a", fg="white",
                font=("Arial", 16, "bold")).pack(pady=15)

        tk.Label(self.main, text="Guest ID", fg="white", bg="#0f172a").pack()
        id_entry = tk.Entry(self.main)
        id_entry.pack()

        tk.Label(self.main, text="Name", fg="white", bg="#0f172a").pack()
        name_entry = tk.Entry(self.main)
        name_entry.pack()

        tk.Label(self.main, text="Phone", fg="white", bg="#0f172a").pack()
        phone_entry = tk.Entry(self.main)
        phone_entry.pack()

        tk.Label(self.main, text="Email", fg="white", bg="#0f172a").pack()
        email_entry = tk.Entry(self.main)
        email_entry.pack()

        def add():
            try:
                guest = Guest(int(id_entry.get()),
                            name_entry.get(),
                            phone_entry.get(),
                            email_entry.get())
                messagebox.showinfo("Result", self.hotel.add_guest(guest))
            except Exception as e:
                messagebox.showerror("Error", str(e))

        def delete():
            messagebox.showinfo("Result",
                                self.hotel.delete_guest(int(id_entry.get())))

        def update():
            messagebox.showinfo("Result",
                                self.hotel.update_guest(
                                    int(id_entry.get()),
                                    name_entry.get(),
                                    phone_entry.get(),
                                    email_entry.get()))

        def search():
            messagebox.showinfo("Result",
                                self.hotel.search_guest(int(id_entry.get())))

        tk.Button(self.main, text="Add Guest", bg="#22c55e", fg="white", command=add).pack(pady=4)
        tk.Button(self.main, text="Delete Guest", bg="#ef4444", fg="white", command=delete).pack(pady=4)
        tk.Button(self.main, text="Update Guest", bg="#3b82f6", fg="white", command=update).pack(pady=4)
        tk.Button(self.main, text="Search Guest", bg="#eab308", fg="black", command=search).pack(pady=4)

    # =========================================================
    #                      RESERVATIONS
    # =========================================================
    def show_reservations(self):
        self.clear()

        tk.Label(self.main, text="📅 Reservations",
                bg="#0f172a", fg="white",
                font=("Arial", 16, "bold")).pack(pady=15)

        rid = tk.Entry(self.main)
        rid.pack()
        add_placeholder(rid, "Reservation ID")

        gid = tk.Entry(self.main)
        gid.pack()
        add_placeholder(gid, "Guest ID")

        room = tk.Entry(self.main)
        room.pack()
        add_placeholder(room, "Room ID")

        cin = tk.Entry(self.main)
        cin.pack()
        add_placeholder(cin, "Check-in YYYY-MM-DD")

        cout = tk.Entry(self.main)
        cout.pack()
        add_placeholder(cout, "Check-out YYYY-MM-DD")

        def make():
            try:
                res = Reservation(
                    int(rid.get()),
                    Guest(int(gid.get()), "temp", "01111111111", "a@a.com"),
                    Room(int(room.get()), "Single", 100),
                    datetime.strptime(cin.get(), "%Y-%m-%d").date(),
                    datetime.strptime(cout.get(), "%Y-%m-%d").date()
                )
                messagebox.showinfo("Result", self.hotel.make_reservation(res))
            except Exception as e:
                messagebox.showerror("Error", str(e))

        def confirm():
            messagebox.showinfo("Result",
                                self.hotel.update_reservation(int(rid.get()),
                                                            new_status="confirmed"))

        def cancel():
            messagebox.showinfo("Result",
                                self.hotel.update_reservation(int(rid.get()),
                                                            new_status="cancelled"))

        def search():
            messagebox.showinfo("Result",
                                self.hotel.search_reservation(int(rid.get())))

        tk.Button(self.main, text="Make Reservation", bg="#22c55e", fg="white", command=make).pack(pady=4)
        tk.Button(self.main, text="Confirm", bg="#3b82f6", fg="white", command=confirm).pack(pady=4)
        tk.Button(self.main, text="Cancel", bg="#ef4444", fg="white", command=cancel).pack(pady=4)
        tk.Button(self.main, text="Search", bg="#eab308", fg="black", command=search).pack(pady=4)
```

```python
root = tk.Tk()
app = Dashboard(root, hotel)
root.mainloop()
```
